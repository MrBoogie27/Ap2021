```python
import json
```

```python
def get_percent_alljson(input, output, percent):
  with open(input, 'r') as file_data, open(output, 'w') as file_new_data:
    obj = json.load(file_data)
    new_len = (len(obj) * percent) // 100
    json.dump(obj[:new_len], file_new_data)
    print(new_len)
```

```python
databases = []
with open('editsql/data/cosql/tables100.json', 'r') as f, open('editsql/data/cosql/tables.json', 'w') as file_new_data:
  obj = json.load(f)
  for el in obj:
    if el['db_id'] in set_db:
      databases.append(el)
  print(len(databases))
  json.dump(databases, file_new_data)
```
